```python
import os
import xml.etree.ElementTree
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import numpy as np
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler  
from sklearn.naive_bayes import GaussianNB 
import sklearn.preprocessing
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
import re

```

```python
def preprocessing(text):
    stop_words = set(stopwords.words('english'))
    tokenisasi = word_tokenize(text)
    filtered = [w for w in tokenisasi if not w in stop_words]
    filtered = []
    stemm =PorterStemmer()
    for w in tokenisasi:
        if w not in stop_words:
            filtered.append(stemm.stem(w))
    
    
    return filtered
```

```python
def read_file(path):
    xep = xml.etree.ElementTree.parse(path).getroot()
    for te in xep.find('text'):
        return preprocessing(te.text.lower())
        #return te.text
```

```python
def import_text(data):
    training = []
    for filename in os.listdir(data):
        training.append(read_file(data + filename))
    return training
```

```python
def gabung(data):
    text = []
    for j in data:
        spasi = ''
        for i in j:
            spasi = spasi + ' ' + i
        text.append(spasi)
    return text
```

```python
path = 'dataset tugas/Training set/'
train = gabung(import_text(path))
train
```

```python
path = 'dataset tugas/Testing set/'
test = gabung(import_text(path))
test
```

```python
data_train = pd.read_csv('dataset tugas/label/Training set.txt', delimiter= ' ',sep="\t", header=None)
lab_train = np.asarray(data_train[1])
lab_train
```

```python
data_tes = pd.read_csv('dataset tugas/label/Testing set.txt', delimiter= ' ',sep="\t", header=None)
lab_tes = np.asarray(data_tes[1])
lab_train
```

# Naive Bayes

```python
from sklearn.pipeline import Pipeline
klasifikasi = Pipeline([('CV', CountVectorizer()),
                      ('tfidf', TfidfTransformer()),
                      ('class', MultinomialNB()),
])
klasifikasi = klasifikasi.fit(train, lab_train)

pred_NB = klasifikasi.predict(test)

pred_NB
```

# KNN

```python
from sklearn.pipeline import Pipeline
klasifikasi = Pipeline([('CV', CountVectorizer()),
                      ('tfidf', TfidfTransformer()),
                      ('class', KNeighborsClassifier(n_neighbors=17, n_jobs=-1)),
])
klasifikasi = klasifikasi.fit(train, lab_train)

pred_KNN = klasifikasi.predict(test)
pred_KNN
```

# Akurasi

```python
akurasi = metrics.accuracy_score(lab_tes, pred_NB)*100
print("Akurasi Naive Bayes = "+str(akurasi))
    
```

```python
akurasi = metrics.accuracy_score(lab_tes, pred_KNN)*100
print("Akurasi KNN = "+str(akurasi))
```

# Presisi

```python
presisiNB = metrics.precision_score(lab_tes, pred_NB, average='micro')*100
print("Presisi Naive Bayes: ", presisiNB)
```

```python
presisiKNN = metrics.precision_score(lab_tes, pred_KNN, average='micro')*100
print("Presisi KNN: ", presisiKNN)
```

# Recall

```python
recallNB = metrics.recall_score(lab_tes, pred_NB, average='micro')*100
print("Recall Naive Bayes: ", recallNB)
```

```python
recallKNN = metrics.recall_score(lab_tes, pred_KNN, average='micro')*100
print("Recall KNN: ", recallKNN)
```

# F1

```python
f1NB = metrics.f1_score(lab_tes, pred_NB, average='micro')*100
print("f1 Naive Bayes: ", f1NB)
```

```python
f1KNN = metrics.f1_score(lab_tes, pred_KNN, average='micro')*100
print("f1 KNN: ", f1KNN)
```
